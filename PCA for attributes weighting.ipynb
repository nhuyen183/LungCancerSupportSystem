```python
GENDER = "GENDER"
AGE = "AGE"
SMOKING = "SMOKING"
LUNG_CANCER = "LUNG_CANCER"
CHRONIC_DISEASE = "CHRONIC DISEASE"
data_df = pd.read_csv("survey lung cancer.csv")

for col in data_df.columns:
    if col != AGE:
        data_df[col] = data_df[col].astype('category').cat.codes
data_df.head(5)
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
df = StandardScaler().fit_transform(data_df)

pca = PCA(n_components=15)
principalComponents = pca.fit_transform(df)
```

```python
df = pd.DataFrame(principalComponents)
df.columns=['G', 'AGE', 'S', 'YF', 'AXIETY', 'PP', 'CD', 'FATIGUE',
           'ALLERGY', 'WHEEZING', 'AC', 'COUGHING', 'SOB', 'SD', 'CP']
df
```

```python
# STEP 3:
cov = pd.DataFrame(df.cov())

from scipy import linalg
# eigenvector and eigenvalues
eig_values, eig_vectors = np.linalg.eigh(cov)

for i in range(len(eig_values)):
    eigvec_cov = eig_vectors[:,i].reshape(1,15).T
    
    print('Eigenvector {}: \n{}'.format(i+1, eig_vectors[i]))
    print('Eigenvalue {} from covariance matrix: {}'.format(i+1, eig_values[i]))
```

```python
# STEP 4:

#sort the eigenvalues in descending order
sorted_index = np.argsort(eig_values)[::-1]
 
sorted_eigenvalue = eig_values[sorted_index]
print('sorted eigenvalues:',sorted_eigenvalue)
#similarly sort the eigenvectors 
sorted_eigenvectors = eig_vectors[:,sorted_index]
print('sorted eigenvectors:',sorted_eigenvectors)
```
